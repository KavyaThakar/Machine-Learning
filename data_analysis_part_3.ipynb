```python
#Liner Regression
def hypothesis(theta_array,x):
  return theta_array[0]+theta_array[1]*x
```

```python
def cost_function(theta_array,x,y,m):
  sum_of_errors=0
  for i in range(m):
    sum_of_errors+=(hypothesis(theta_array,x.iloc[i,0])-y.iloc[i,0])**2
  return sum_of_errors/(2*m)
```

```python
#gradient descent formulas
def gradient_descent(theta_array,x,y,learning_rate,m):
  dtheta_0=0
  dtheta_1=0
  for i in range(m):
    dtheta_0+=(hypothesis(theta_array,x[i])-y[i])
    dtheta_1+=(hypothesis(theta_array,x[i])-y[i])*x[i]
  theta_array[0]=theta_array[0]-learning_rate*(dtheta_0/m)
  theta_array[1]=theta_array[1]-learning_rate*(dtheta_1/m)
  return theta_array
```

```python
def training(x,y,learning_rate,iterations):
  theta0=0
  theta1=0
  cost_values=[]
  theta_array=[theta0,theta1]
  m=x.size
  for i in range(iterations):
    theta_array=gradient_descent(theta_array,x,y,learning_rate,m)
    cost_values.append(cost_function(theta_array,x,y,m))
    y_predict=x*theta_array[1]+theta_array[0]
    plt.plot(x,y_predict,color='pink')
    plt.scatter(x,y)
    plt.show()
  return cost_values
```
